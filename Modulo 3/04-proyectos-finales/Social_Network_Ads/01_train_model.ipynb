```python
import opendatasets as od
import pandas as pd
import numpy as np
import os
```

```python
os.environ["KAGGLE_CONFIG_DIR"] = "kaggle.json"

od.download("https://www.kaggle.com/datasets/rakeshrau/social-network-ads")
```

```python
df = pd.read_csv("./social-network-ads/Social_Network_Ads.csv")
df.head()
```

```python
conteo= df["Gender"].value_counts()
print(conteo)
```

```python
df.info()
```

```python
#cambiar Grender a 0 y 1
df["Gender"] = df["Gender"].map({"Male": 0, "Female": 1})
df.head()
```

```python
df.info()
```

```python
x = df.drop(["User ID", "Purchased"], axis=1)
y = df["Purchased"]
```

```python
x.head()
```

```python
y.head()
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.21, random_state=42)

print(x_train.shape)
print(y_train.shape)
print(x_test.shape)
print(y_test.shape)
```

```python
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LogisticRegression

pipeline = Pipeline([
    ("scaler", StandardScaler()),
    ("classifier", LogisticRegression(random_state=42))
])

pipeline
```

```python
pipeline.fit(x_train, y_train)
```

```python
from sklearn.metrics import accuracy_score, precision_score

y_pred = pipeline.predict(x_test)

y_pred_proba = pipeline.predict_proba(x_test)[:, 1]

precision = accuracy_score(y_test, y_pred)

print(f"Precision: {precision * 100}")

precision = precision_score(y_test, y_pred)

```

```python
#Guardamos el modelo

import pickle

with open("model.pkl", "wb") as f:
    pickle.dump(pipeline, f)
```
